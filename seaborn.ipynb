```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
print(sns.get_dataset_names())
```

```python
#load a build dataset 
crash=sns.load_dataset('car_crashes')
crash.head()
```

# Distribution Plots

# Distribution Plot

```python
sns.displot(crash['not_distracted'],bins=25)



plt.show()
```

# Joint Plot 

```python
sns.jointplot(x='speeding',y='alcohol',data=crash,kind='reg')

plt.show()
```

# KDE Plot

```python
sns.kdeplot(data=crash,x='alcohol',fill=True)
```

# Pair Plot 

```python
#h=crash.select_dtypes(include='number').head()


#print(crash.select_dtypes(include='number').head())

sns.pairplot(crash)
#h
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.pairplot(tips,hue='sex',palette='viridis',diag_kind='kde',corner=True)
```

# Rug Plots

```python
# sns.set_style('darkgrid')
sns.set_context('paper',font_scale=1.5)

sns.rugplot(tips['tip'],color='red')
```

# Categorical Plots

# Bar Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.median)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.cov)
```

# Count Plot

```python
sns.countplot(x='smoker',data=tips)
```

# Box Plot

```python
plt.figure(figsize=(15,9))
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')

plt.legend(loc=0)
```

# Violin Plot 

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

# Strip Plot

```python
plt.figure(figsize=(14,10))

sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',
              jitter=True,
              dodge=True
             )
```

# Swarm Plot

```python

sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='white')
```

# Matrix Plots

# HeatMaps

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

crashmx=crash.corr()
sns.heatmap(crashmx,annot=True,cmap='Blues')
```

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

flights=sns.load_dataset('flights')
flights=flights.pivot_table(index='month',columns='year',values='passengers')
flights

sns.heatmap(flights,cmap='Blues',linecolor='white',linewidth=2)
```

# Cluster Map

```python
iris=sns.load_dataset('iris')
iris
iris.pop("species")
sns.clustermap(iris,cmap='Blues',standard_scale=1)
```

```python
sns.clustermap(flights,cmap="Blues", standard_scale=1)
```

# Pair Grid

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)
sns.set_style('dark')

iris=sns.load_dataset('iris')
iris_g=sns.PairGrid(iris,hue='species',
             x_vars=["sepal_length","sepal_width"],
             y_vars=["petal_length","petal_width"]
             )
iris_g.map(plt.scatter)
iris_g.add_legend()
```

# Facet Grid

```python
attention=sns.load_dataset('attention')
attention

att_fg=sns.FacetGrid(attention,col="subject",col_wrap=5,height=1.5)
att_fg.map(plt.plot,'solutions','score',marker='*')
```

# Regression Plots

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_context('paper', font_scale=1.4)

plt.figure(figsize=(8,6))

sns.lmplot(x='total_bill',y='tip',data=tips,
           hue='sex',
           markers=['o','^'],
           scatter_kws={'s':100,'linewidth':0.5,'edgecolor':'w'},
           col='day',
           height=8,
           aspect=0.6
           
          )
```
