# Making Changes To The Data

```python
ex=pd.read_csv('pokemon_data.csv')
ex['Total']=ex['HP']+ex['Attack']+ex['Defense']+ex['Sp. Atk']+ex['Sp. Def']+ex['Speed']
ex = ex.drop(columns=['Total'])
ex['Total']=ex.iloc[:,4:10].sum(axis=1)
cols=list(ex.columns)
print(cols)
ex=ex[cols[0:4]+[cols[-1]]+cols[4:-1]]
ex.head(5)
```

# Saving our Data (Exporting into Desired Format

```python
ex.to_csv('modified.csv',index=False)
ex.to_excel('modified.xlsx',index=False)
ex.to_csv('modified.txt',index=False,sep='\t')
```

# Working with large amounts of data

```python
new_df=pd.DataFrame(columns=ex.columns)
for ex in pd.read_csv('modified.csv',chunksize=5):
    results = ex.groupby(['Type 1']).count()
    new_df = pd.concat([new_df, results])
    print(new_df)
```
